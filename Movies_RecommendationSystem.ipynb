```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

            Above Imp points 

1. Numpy and Pandas: Imported for handling arrays and data.
2. Difflib: Imported for comparing sequences (e.g., strings).
3. TfidfVectorizer: Converts text data into numerical form using TF-IDF (word importance).
4. Cosine Similarity: Measures similarity between two pieces of text.


### DATA COLLECTION AND PRE-PROCESSING


```python
#Loading the data from the csv file to apandas dataframe
movies_data = pd.read_csv('movies.csv')
```

```python
#Printing the first 5 rows of the dataframe
movies_data.head()
```

```python
#No.of rows and coloumns in the data frame
movies_data.shape
```

```python
#selecting the relevant features for recomendation
selected_features = ['genres' , 'keywords' ,'tagline','cast','director' ]
print(selected_features)
```

```python
#Replacing the null values with null string
for feature in selected_features:
    movies_data[feature] = movies_data[feature].fillna('')
```

```python
#Combiniing all the selected features
combined_features = movies_data['genres'] +' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director'] 
```

```python
print(combined_features)
```

```python
# converting the text data to features vectors

vectorizer = TfidfVectorizer()
```

```python
feature_vector = vectorizer.fit_transform(combined_features) 
```

```python
print(feature_vector)
```

Cosine similarity


```python
#Getting the similarity scores using cosine similarity

similarty = cosine_similarity(feature_vector)
```

```python
print(similarty)
```

```python
print(similarty.shape)
```

```python
#getting the movie name from the user

movie_name  = input('Enter your favourite Movie : ')
```

```python
#creating a list with all the movies names given in the dataset

list_of_all_titles = movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
# finding the close match for the movie name given by the user
find_close_match = difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_close_match)
```

```python
close_match =find_close_match[0]
print(close_match)
```

```python
#find the index of the movie with title
index_of_the_movies = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_the_movies)
```

```python
#getting a lsit of similar movies
similarity_score = list(enumerate(similarty[index_of_the_movies]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#Sorting the movie based on the similarity score
sorted_similar_movies = sorted(similarity_score,key = lambda x:x[1],reverse = True)
print(sorted_similar_movies)
```

```python
#print the name of similar movies based on the index

print('Movies suggested for : \n')

i = 1

for movies in sorted_similar_movies:
    index = movies[0]
    title_from_index = movies_data[movies_data.index==index]['title'].values[0]
    if (i<=30):
        print(i, '.',title_from_index)
        i+=1
```
